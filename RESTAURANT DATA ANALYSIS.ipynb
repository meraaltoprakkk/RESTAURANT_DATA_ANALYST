```python
# KUTU GRAFİĞİNİN OLUŞTURULMASI (BOX PLOT)
```

**sayısal değişkenın dağılımı incelenmek istiyorsa, histogramla beraber kutu grafıgıne de bakılmalıdır. Kutu Grafıgı = Box Plot**

```python
import seaborn as sns
tips= sns.load_dataset("tips")
df= tips.copy()
df.head()
```

```python
#yemegın toplam değerine ait ifadeler :
sns.boxplot(x=df["total_bill"]);
```

```python
# VİOLİN :#Dağılım anlamında bıze bılgı sunar. yogunluk grafıgı ıle kutu grafıgının kesısımı olarak elımızde bulunabılır.
```

```python
df.head()
```

```python
#odenen para ile ilgili olan medyanı ıle ılgılı olan bır grafık. ortadakı cızgı medyanı ifade eder. Vıolın grafık tercıhen kullanılır.
sns.catplot(y="total_bill",kind="violin", data=df);
```

# VIOLIN GRAFIK ÇAPRAZLAMALAR

```python
#Gunler boyutunu ekledık. gunler ve odenen hesap anlamında bır ekleme yapıldı. kategorık değişken gunlerımız. kategorık değişken etkısıyle
#sayısal degıskenımız olan hesabı da eklemıs olduk. 
sns.catplot(x="day", y="total_bill", kind = "violin" , data=df);
```

```python
#gunlere göre, odenen hesabana gore, cınsıyet kırılımına bakmak. 
sns.catplot(x="day", y="total_bill", hue= "sex", kind = "violin" , data=df);
```

**Not:Elimizde tek bir değişken varsa bu değişkenin tipine bakıyoruz. Bu değişken kategorık ıse sütun grafıgı kullandık barplot. 
 Eğer bu değişken sayısal ise ( sürekli/ kesikli) bu değişkeni görselleştirmek için histogram, yogunluk, box plot ve vıolın grafık kullandık.
Bunlara eklemiş olunan ikinci üçüncü değişken ise boyut anlamında çaprazlama anlamında yapılan eklemelerdir.
Şimdi ise tek değişkenden iki üç değişkene geçmiş olduk.**

# Korelasyon Grafikleri : SCATTERPLOT 
Değişkenler arası ilişkiye korelasyon denir. 

## Scatterplot :İki değişken arasındaki ilişkiyi ifade eden yaklasım scatterplot yaklasımıdır. Scatterplot bıze ; sayısal değişken arasındaki 
## ilişkiyi gösteririr. 2 degısken birarada değerlendırıldıgınde verı setının yapısı hakkında bilgi veriri
total_bill : yemeğin toplam fiyatı ( bahşiş ve vergi dahil)

tip : bahşiş

sex: ücreti ödeyen kişinin cinsiyeti ( 0 = male, 1 = female)

smoker : grupta sigara içen var mı ? ( 0= No, 1 = Yes)

day : gün (3=thur, 4= fri, 5= sat, 6 = sun)


time : ne zaman? ( 0 = day, 1 = night )

size = grupta kaç kişi var ?


```python
import seaborn as sns
tips= sns.load_dataset("tips")
df= tips.copy()
df.head()
```

```python
#2 sayısaldeğişken. dikkat et kategorık degısken degıl!!!!!!!!!!!!! korelasyon analızınde boyledır.
sns.scatterplot(x= "total_bill" , y= "tip", data= df);
```

Ödenen hesap ve ödenen bahşiş arasındaki ilişki görselleştirilmiştir.
Saçılım grafıgı bıze 2 degısken arasındakı ilişkiyi göstermiş. toplam
hesap arttıkça ödenen bahşiş artmıştır.

# Korelasyon Çaprazlamalar

```python
sns.scatterplot(x= "total_bill" , y= "tip", hue="time", data= df);
```

ödenen hesap arttıkça ; verilen bahşiş artıyor, akşam yemeklerinde ödenen hesap arttıkça bahşiş artmıstır daha çok.

```python
sns.scatterplot(x= "total_bill" , y= "tip", hue="time", style="time",  data= df);
```

```python
sns.scatterplot(x= "total_bill" , y= "tip", hue="day", style= "time" , data= df);
```

```python
sns.scatterplot(x= "total_bill" , y= "tip", hue="day", style= "time" , data= df);
```

Ödenen hesap ile verilen bahşiş ilişkine bakıldı, fakat günler kırılımında yapıldı bu. burada günler ve zaman kıvrım olarak verıldı.
günler ve zaman ıkı boyut oldu.

**Sürekli değişkeni boyut olarak eklersek ;** 
**yanı kategorık değil de sayısal bir değişken eklemek istersem ;**
**Sayısal bir değişken eklemek için size argumanını kullanırız.**

```python
sns.scatterplot(x= "total_bill" , y= "tip", hue="size", size="size" , data= df);
```

hue argümanı renk kodlaması için veri çerçevesindeki hangi sütunun kullanılması gerektiğini belirler.

# Doğrusal İlişkinin Gösterilmesi 

İki sayısal değişkenin arasındaki ilişkiyi bir grafık aracılıgıyla sunmuştu.
Bir dogru aracılıgıyla olan ılıskıye bakalım.

```python
import seaborn as sns
tips= sns.load_dataset("tips")
df= tips.copy()
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x="total_bill", y="tip", data= df);
```

2 değişken arasındaki doğrusal ilişki. Doğru etrafındakı acık renklı mavıler sapmaları ifade edıyor.

```python
sns.lmplot(x="total_bill", y="tip", hue= "smoker", data= df);
```

```python
sns.lmplot(x="total_bill", y="tip", hue= "smoker", col ="time", data= df);
```

```python
sns.lmplot(x="total_bill", y="tip", hue= "smoker", col ="time", row="sex", data= df);
```
